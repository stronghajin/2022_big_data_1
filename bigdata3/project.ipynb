```python
def extract_noun_from_kiwi_analyzed(comment):
    try:
        kiwi_analyzed_comment = kiwi.analyze(comment)
    except ValueError:
        return []

    return [
        token.form 
        for token 
        in kiwi_analyzed_comment[0][0] 
        if token.tag.startswith('N') and token.len > 1
    ]
```
